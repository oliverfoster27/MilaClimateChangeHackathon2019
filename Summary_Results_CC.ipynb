```python
column_average = []
columns = []
for col in df.columns:
    value = df['%s'%(col)].mean()
    column_average.append(value)
    
    columns.append(col)
```

```python
df_averageSum = pd.DataFrame()
df_averageSum['features'] = columns
df_averageSum['column_average'] = column_average
```
