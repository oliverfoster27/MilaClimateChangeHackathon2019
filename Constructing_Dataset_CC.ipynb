### Groupby section for dataset - Aggregating the dataset on date and grid location ### 

```python
df = pd.read_csv('D:\\Climate Change Hackathon Data\\MilaClimateChangeHackathon2019-master\\MilaClimateChangeHackathon2019-master\\data\\Accidents_2012_2017\\Accident_ResponseData_WithGrids1.csv',index_col=0)
```

```python
df.sort_values('Date of collision',ascending=True).reset_index(drop=True)
```

```python
df['Date of collision'] = df['Date of collision'].apply(pd.to_datetime)
```

```python
#(df.groupby('Date of collision').count())
```

```python
#df['Date of collision'].value_counts()
```

```python
groupby_list = ['Date of collision','Grid']
temp = df.groupby(groupby_list).agg({
    'Date of collision' : [
        ('Number of incidents', 'count')],
    'Implied vehicules' : [
        ('Implied vehicules', 'sum')],
    'Serious injury':[('Serious injury','sum')],
    'Light injury':[('Light injury','sum')],
    'Total victims':[('Total victims','sum')],
     'Trucks and car':[( 'Trucks and car','sum')],
    'Heavy tractor':[('Heavy tractor','sum')],
    'Tool equipment':[('Tool equipment','sum')],
    'Bus':[('Bus','sum')],
    'Bicycle':[('Bicycle','sum')],
    'Moped':[('Moped','sum')],
    'Motorcycle':[('Motorcycle','sum')],
    'Emergency vehicule':[('Emergency vehicule','sum')],
    'Snowmobile':[('Snowmobile','sum')],
    'VHR':[('VHR','sum')],
    'Other types':[('Other types','sum')],
    'Non-Precise vehicule':[('Non-Precise vehicule','sum')],
    'Pedestrian death':[('Pedestrian death','sum')],
    'Pedestrian injury':[('Pedestrian injury','sum')],
    'Pedestrian victims':[('Pedestrian victims','sum')],
    'Motorcycle death':[('Motorcycle death','sum')],
    'Motorcycle injuries':[('Motorcycle injuries','sum')],
    'Motorcycle victims':[('Motorcycle victims','sum')],
    'Bicycle death':[('Bicycle death','sum')],
    'Bicycle injury':[('Bicycle injury','sum')],
    'Bicycle victims':[('Bicycle victims','sum')],
    'Non-Precise vehicule':[('Non-Precise vehicule','sum')],
    'Severity: Inferior damages':[('Severity: Inferior damages','sum')],
    'Severity: Material damages':[('Severity: Material damages','sum')],
    'Severity: Serious injury':[('Severity: Serious injury','sum')],
    'Severity: Light injury':[('Severity: Light injury','sum')],
    'Severity: Death':[('Severity: Death','sum')]
})
```
