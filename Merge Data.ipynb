```python
import pandas as pd
```

```python
prec_df = pd.read_csv(r"C:\Users\olive\Documents\GitHub\MilaClimateChangeHackathon2019\data\prepared_precipitation_data.csv")
prec_df['time'] = pd.to_datetime(prec_df['time'])
prec_df['time'] = [x.date() for x in prec_df['time'].tolist()]
prec_df = prec_df.reset_index(drop=True)
prec_df = prec_df.sort_values(['time', 'Grid']).set_index(['time','Grid'])
```

```python
acc_df = pd.read_csv(r"C:\Users\olive\Documents\GitHub\MilaClimateChangeHackathon2019\data\AccidentData_GroupBy.csv")
acc_df['Grid'] = [x.replace('GRID_', '') for x in acc_df['Grid'].tolist()]
acc_df = acc_df.rename({'Date of collision': 'time'}, axis=1)
acc_df['time'] = pd.to_datetime(acc_df['time'])
acc_df['time'] = [x.date() for x in acc_df['time'].tolist()]
acc_df = acc_df.reset_index(drop=True)
acc_df = acc_df.sort_values(['time', 'Grid']).set_index(['time', 'Grid'])
```

```python
merged_df = pd.concat([acc_df, prec_df], axis=1).fillna(0)
```

```python
merged_df = merged_df.drop(['lat', 'lon'], axis=1)
merged_df = merged_df.reset_index(drop=True)
```
